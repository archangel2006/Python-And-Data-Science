```python
# REGULAR EXPRESSIONS
        # A Regular Expression or RegEx is a special sequence of characters that uses a search pattern to find a string or set of strings. 
   # It can detect the presence or absence of a text by matching it with a particular pattern and also can split a pattern into one or more sub-patterns. 
        # Python has a built-in module named “re” that is used for regular expressions in Python. 

import re
s='Hello! How are you?'
match=re.search(r'are',s)
print(f'Start index: {match.start()}')
print(f'End index: {match.end()}')



# Matching patterns within strings using various functions and metacharacters.

# Compiling patterns for efficient reuse in multiple operations.
```

```python
# METACHARACTER (special character) - characters with special meanings

# \ (Backslash) - Escapes special characters.
s='Hello! Hi.'
match=re.search(r'\.',s)
print(match)

# [ ] (Square Brackets) - Represents a set of characters.
s='The quick brown fox jumps over the lazy dog'
pattern1='[a-c]'
pattern2='[efg]'
result1=re.findall(pattern1,s)   # re,findall(pattern,string) returns a list of all such characters
result2=re.findall(pattern2,s)
print(result1,result2)
print()

# ^ (Caret) - Matches the start of a string.
regex=r'^The'
strings = ['The quick brown fox', 'The lazy dog', 'A quick brown fox'] 
for st in strings:
    if re.match(regex,st):
        print(f'Matched: {st}')
    else:
        print(f'Not Matched: {st}')
print()

# $ (Dollar) - Matches the end of a string.
regex=r'fox$'
strings = ['The quick brown fox', 'The lazy dog', 'A quick brown fox'] 
for st in strings:
    if re.search(regex,st):
        print(f'Matched: {st}')
    else:
        print(f'Not Matched: {st}')
print()


# . (Dot) - Matches any single character except newline.

string = "The quick brown fox jumps over the lazy dog."
pattern=r'brown fox'
match=re.search(pattern,string)
if match: 
    print("Match found!") 
else: 
    print("Match not found.") 
```

```python
# SPECIAL SEQUENCES - Shortcuts for common patterns 

# \A - matches if string begins with character
st= 'Start < Hello 143 Bye > End'
pattern=r'\AStart'
print(re.search(pattern,st))

# \b - matches if str begins or ends with character
st= 'Start < Hello 143 Bye > End'
pattern1=r'Start\b'
pattern2=r'\bEnd'
print(re.search(pattern1,st))
print(re.search(pattern2,st))

# \B - opposite of \b
st= 'Start < Hello143 Bye > End'
pattern1=r'Start\b'
pattern2=r'\bEnd'
print(re.search(pattern1,st))
print(re.search(pattern2,st))

# \d - matches digits [0-9] 
st= 'Start < Hello 143 Bye > End'
pattern=r'\d'
print(re.search(pattern,st))

# \D - matches non digit [^0-9]
st= 'Start < Hello 143 Bye > End'
pattern=r'\D'
print(re.search(pattern,st))

# \s - matches any white space character
st= 'Start < Hello 143 Bye > End'
pattern=r'\s'
print(re.search(pattern,st))

# \S - matches any non whitespace character
st= 'Start < Hello 143 Bye > End'
pattern=r'\S'
print(re.search(pattern,st))

# \w - matches any alphanumeric characters [a-z A-Z 0-9]
st= 'Start < Hello 143 Bye > End'
pattern=r'\w'
print(re.search(pattern,st))

# \W - matches any non alphanumeric character 
st= 'Start < Hello 143 Bye > End'
pattern=r'\W'
print(re.search(pattern,st))

# \Z - matches if string ends with character
st= 'Start < Hello 143 Bye > End'
pattern=r'End\Z'
print(re.search(pattern,st))
```
