```python
for i in range (7,48):
    if i+3==50:
        print(50)
    else:
        print(i,end=',')
        print(i+3,end=',')
```

```python
for i in range(20,81):
    if i-1==79:
        print(79)
    else:
        print(i,end=',')
        print(i-1,end=',')
```

```python
n=53
for i in range(53,100):
    print(f'{n},{n},{n-13},{n-13}',end=',')
    n=n-13
```

```python
n=14
k=1
while n+6*k<100:
    print(f'{n},{2*n}',end=',')
    n=n+6*k
    k=k+1
```
