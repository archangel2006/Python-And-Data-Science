```python
# MATPLOTLIB
        # Matplotlib is a low level graph plotting library in python that serves as a visualization utility. 
        # Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias: 
```

```python
# PLOTTING X and Y POINTS - The plot() function is used to draw points (markers) in a diagram. 
    # By default, the plot() function draws a line from point to point. 

# example1- plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 2, 6, 8])    # Parameter 1 is an array containing the points on the x-axis. 
y_points = np.array([3, 8, 1, 10])  # Parameter 2 is an array containing the points on the y-axis. 

plt.plot(x_points,y_points,'c',marker='*',ms=20,linewidth='2',mfc='g',linestyle='dotted')
plt.show()
```

```python
#example2
import numpy as np 
import matplotlib.pyplot as plt 

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125]) 
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330]) 

plt.xlabel("Average Pulse") 
plt.ylabel("Calorie Burnage") 
plt.title("Sports Watch Data")
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#example3 - multiple plots

#plot 1: 
x = np.array([0, 1, 2, 3]) 
y = np.array([3, 8, 1, 10])
plt.title('plot1')
plt.subplot(1, 2, 1) 
plt.plot(x,y,color='y')

#plot 2: 
x = np.array([0, 1, 2, 3]) 
y = np.array([10, 20, 30, 40]) 
plt.title('plot2')
plt.subplot(1, 2, 2) 
plt.plot(x,y,color='r') 

plt.show()


# The subplot() function takes three arguments that describes the layout of the figure. 
    # The layout is organized in rows and columns, which are represented by the first and second argument.
    # The third argument represents the index of the current plot. 
#plt.subplot(1, 2, 1) 
    #the figure has 1 row, 2 columns, and this plot is the first plot. 
```

```python
# BAR GRAPHS

import matplotlib.pyplot as plt 
import numpy as np 
x = np.array(["A", "B", "C", "D"]) 
y = np.array([3, 8, 1, 10]) 
plt.bar(x,y,color=['r','y','g','m']) 
plt.show() 

import matplotlib.pyplot as plt 
import numpy as np 
x = np.array(["A", "B", "C", "D"]) 
y = np.array([3, 8, 1, 10]) 
plt.barh(x,y,color=['r','y','g','m']) 
plt.show() 
```

```python
# HISTOGRAM

import matplotlib.pyplot as plt 
import numpy as np 
x = np.random.normal(170, 10, 250) # generates a random result
plt.hist(x,color=['c']) 
plt.show() 
```

```python
# PIE CHARTS

import matplotlib.pyplot as plt 
import numpy as np 
y = np.array([35, 25, 25, 15]) 
mylabels = ["Apples", "Bananas", "Guavas", "Cherries"] 
myexplode = [0.2, 0, 0, 0]      # Pull the "Apples" wedge 0.2 from the center of the pie
mycolors = ["black", "hotpink", "c", "#4CAF50"] 
plt.pie(y, labels = mylabels,startangle=90,colors=mycolors,explode=myexplode)
plt.legend(title = "Fruits:") 
plt.show()   
```
