```python
# DATA VISUALIZATION IN PYTHON
       # Data visualization helps us explore and understand data by representing it graphically. 
       # In Python, libraries like Matplotlib, Seaborn, and Pandas make it easy to create various visualizations.

```

```python
# UNDERSTANDING DATA VISUALIZATION
Data visualization involves creating graphical representations of data. It helps in:

Understanding data patterns.
Identifying trends, anomalies, and distributions.
Communicating findings effectively.

Python libraries for visualization:
Matplotlib: Base library for creating visualizations.
Seaborn: Built on Matplotlib, provides a high-level interface.
Pandas: Direct plotting methods for DataFrames.
```

```python
# CREATING VISUALIZATIONS

# 1. BAR CHARTS - used to compare categorical data
import matplotlib.pyplot as plt

categories = ['A','B','C']
values= [10,15,7]

plt.bar(categories,values,color=('red','pink','blue'))
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# 2. LINE PLOTS - used to show trends over time or sequence

import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[3,6,12,9]
plt.plot(x,y,marker='*',linestyle='--',color='cyan')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# 3. AREA PLOTS - similar to line plots but with the area under the curve filled

import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[3,6,12,9]
plt.fill_between(x,y,color='lightblue',alpha=0.5)
plt.title('Area Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# 4. HISTOGRAMS - shows the distribution of numerical data

import numpy as np
import matplotlib.pyplot as plt

data=np.random.randn(1000)
plt.hist(data,bins=20,color='orange',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. PIE CHARTS - pie charts represent proportions

import matplotlib.pyplot as plt

sizes=[20,30,50]
labels=['Category1','Category2','Category3']
colors=['pink','lightblue','lightgreen']

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
# 6. BOX PLOTS - shows the distribution of data and highlights outliners

import matplotlib.pyplot as plt

data = [7, 8, 6, 4, 5, 9, 10, 15, 2]
#plt.boxplot(data,vert="True",patch_artist=True,boxdrop=data())
plt.boxplot(data,vert=False)
plt.title('Box Plot')
plt.xlabel('Values')
plt.show()
```

```python
# 7. SCATTER PLOT - show relationships between two variables

import matplotlib.pyplot as plt

x=[3,6,9,12]
y=[5,10,15,12]
plt.scatter(x,y,color=['purple','red','cyan','magenta'])
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# 8. TIME SERIES PLOT -used to display data points over time

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# creating time series
dates=pd.date_range('2024-01-01', periods=10)  # dates from 2024-01-01 to 2024-01-10
values=np.random.randn(10)   # values in between -1 to +1

plt.plot(dates,values,marker='^')
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. FIGURES AND SUBPLOTS -allow multiple plots in a single figure


x=[3,6,9,12]
y=[5,10,15,12]
categories = ['A','B','C']
values= [10,15,7]
data=np.random.randn(5)

import matplotlib.pyplot as plt

fig,ax=plt.subplots(2,2,figsize=(8,6))

ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[1,0].bar(categories,values)
ax[1,1].hist(data,bins=10)

plt.tight_layout()
plt.show()

```

```python
# PLOTTING FUNCTIONS WITH PANDAS

import pandas as pd
df=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df.plot(kind='line')
df.plot(kind='bar')
df.plot(kind='box')
plt.show()
```
